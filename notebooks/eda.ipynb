```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter

df = pd.read_csv('../data/Coursera.csv')

# Preview
print(df.head())
print(df.info())

# Distribusi difficulty
sns.countplot(x='difficulty', data=df)
plt.title("Distribusi Difficulty")
plt.show()

# Distribusi rating
sns.histplot(df['rating'].dropna(), bins=20)
plt.title("Distribusi Rating Kursus")
plt.show()

# Analisis skills
all_skills = []
for s in df['skills'].dropna():
    all_skills.extend([x.strip().lower() for x in s.split(',')])
skill_count = Counter(all_skills)
top_skills = skill_count.most_common(20)
print(top_skills)
```
